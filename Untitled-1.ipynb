```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter
import pandas as pd
```

```python
path="C:\\Users\\Admin\\OneDrive\\Documents\\TASK SATHYA .xls"
df=pd.read_excel(path,skiprows=3)
```

```python
df.head()
```

```python
df_cleaned = df.dropna(subset=['Country Name'])
```

```python
df_long = df_cleaned.melt(id_vars=["Country Name"], 
                          var_name="Year", 
                          value_name="Population")
df_long = df_long.rename(columns={"Country Name": "Country"})
```

```python
df_long = df_long.dropna(subset=["Population"])
```

```python
df_long["Population"] = pd.to_numeric(df_long["Population"], errors='coerce')
```

```python
numeric_columns = df_long.select_dtypes(include='number').columns
```

```python
top_countries = df_long.groupby('Country')[numeric_columns].mean().sort_values(by="Population", ascending=False).head(50).index
```

```python
df_top_countries = df_long[df_long['Country'].isin(top_countries)]
```

```python
df_top_countries = df_top_countries.dropna(subset=["Population"])
```

```python
plt.figure(figsize=(30, 8))
```

```python

plt.figure(figsize=(15, 8))
ax = sns.barplot(data=df_top_countries, x="Population", y="Country", errorbar=None, width=0.8)

plt.title("Top 30 Countries by Population (1960-2023)", fontsize=20)
plt.xlabel("Average Population", fontsize=16)
plt.ylabel("Country", fontsize=16)

for p in ax.patches:
    ax.annotate(f'{int(p.get_width()):,}', 
                (p.get_width(), p.get_y() + p.get_height() / 2), 
                xytext=(5, 0), 
                textcoords='offset points', 
                ha='left', va='center', fontsize=10)


ax.xaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{int(x):,}'))

plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```
